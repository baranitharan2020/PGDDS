```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
product_cat=np.array(['Furniture','IT','Electronics'])
sale=np.array([5090.42,4532.93,4341.98])
```

```python
plt.bar(product_cat,sale)
plt.show()
```

```python
plt.bar(product_cat, sale, width=0.5, align='center',edgecolor='red', color='black', )
plt.show()
```

```python
plt.bar(product_cat, sale, width= 0.5, align='center', edgecolor='Orange',color='cyan')

plt.title('Sales of the Year 2021 \n',fontdict={'color':'blue','fontsize':20,'fontweight':21})
plt.show()
```

```python
plt.bar(product_cat, sale, width= 0.5, align='center', edgecolor='Orange',color='cyan')

plt.title('Sales of the Year 2021 \n',fontdict={'color':'blue','fontsize':20,'fontweight':21})

plt.xlabel('Products',fontdict={'color':'blue','fontsize':20,'fontweight':101})

plt.ylabel('Sales',fontdict={'color':'blue','fontsize':20,'fontweight':101})

plt.show()
```
